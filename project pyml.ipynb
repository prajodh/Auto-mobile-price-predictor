## Initial data cleaning

```python
auto['num-of-doors'].replace('?','four',inplace=True)
```

```python
newauto= auto.replace('?',np.nan)
```

```python
newauto
```

```python

newauto.info()
```

```python
newauto["normalized-losses"] = pd.to_numeric(newauto["normalized-losses"])
newauto["price"] = pd.to_numeric(newauto["price"])
newauto["horsepower"] = pd.to_numeric(newauto["horsepower"])
newauto["peak-rpm"] = pd.to_numeric(newauto["peak-rpm"])
newauto["bore"] = pd.to_numeric(newauto["bore"])
newauto["stroke"] = pd.to_numeric(newauto["stroke"])
```

```python
newauto.info()
```

```python
newauto['normalized-losses']=newauto['normalized-losses'].fillna(value=newauto['normalized-losses'].mean(skipna=True))
newauto['price']=newauto['price'].fillna(value=newauto['price'].mean(skipna=True))
newauto['horsepower']=newauto['horsepower'].fillna(value=newauto['horsepower'].mean(skipna=True))
newauto['peak-rpm']=newauto['peak-rpm'].fillna(value=newauto['peak-rpm'].mean(skipna=True))
newauto['bore']=newauto['bore'].fillna(value=newauto['bore'].mean(skipna=True))
newauto['stroke']=newauto['stroke'].fillna(value=newauto['stroke'].mean(skipna=True))
```
